```python
import numpy as np
from PIL import Image
import cv2 as cv
import h5py

import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

np.set_printoptions(precision=7)
```

```python
def load_dataset():
    dataset = h5py.File('./catvnoncat.h5',"r")
    train_set_x_orig = np.array(dataset["X_train"][:])
    train_set_y_orig = np.array(dataset["y_train"][:])
    
    val_set_x_orig = np.array(dataset["X_val"][:])
    val_set_y_orig = np.array(dataset["y_val"][:])
    
    classes = np.array(dataset["classes"][:])
    
    return train_set_x_orig, train_set_y_orig, val_set_x_orig, val_set_y_orig, classes   
```

```python
x_train_ori, y_train_ori, x_test_ori, y_test_ori, classes = load_dataset()
```

```python
def show_image(data, label):
    fig, ax = plt.subplots(3, 10, figsize=(18,5))
    fig.subplots_adjust(hspace=0.5, wspace=0.1)
    for j in range(0, 3):
        for i in range(0, 10):
            ax[j,i].imshow(data[i+j*10])
            ax[j,i].set_title(classes[label[i+j*10,0]].decode("utf-8"))
            ax[j,i].axis('off')
    plt.show()

show_image(x_test_ori, y_test_ori)
```

```python
x_train = x_train_ori.reshape(x_train_ori.shape[0], -1)
x_test = x_test_ori.reshape(x_test_ori.shape[0], -1)

x_train = x_train/ 255
x_test = x_test/255
```

```python
def open_image(filename, show=True):
    image = Image.open(filename)
    iamge = np.array(image)
    if show:
        plt.imshow(image)
        plt.show
    return image
```

```python
def resize_image(image, shape=(64,64), show=True):
    new_image = cv.resize(image, shape, interpolation=cv.INTER_AREA)
    if show:
        plt.imsow(new_image)
        plt.show()
    return new_image
```

```python
def affine_forward(x, W, b):  
    v = x.dot(W) + b            
    return v

def affine_backward(dout, x, W, b):
    dW = x.T.dot(dout)                                 
    db = np.sum(dout, axis=0, keepdims=True)           
    dx = dout.dot(W.T)                                 
    return dW, db, dx
```

```python
def binary_cross_entropy_loss(predictions, targets, epsilon=+1e-10):
    N = predictions.shape[0]
    log_likelihood = targets* np.log(predictions+epsilon) + (1.-targets)*np.log(1.-predictions+epsilon)
    loss = -np.sum(log_likelihood)/N
    
    return loss

def sigmoid_forward(x):
    out = 1/(1 + np.exp(-x))
    return out
```

```python
def train_one_layer(X, y, W=None, b=None, lr=0.005, epochs=100, verbose=True):
    
    num_train, dim = X.shape
    num_classes = 1
    
    if W is None:
        w = 0.02*np.random.rand(dim, num_classes)
    if b is None:
        b = np.zeros((1, num_classes))
        
    loss_history = []
    
    for ep in range(epochs):
        
        layer1 = affine_forward(X, W, b)
        
        act1 = sigmoid_forward(layer1)
        
        error = act1 - y
        
        loss = binary_cross_entropy_loss(act1, y)
        
        error /= num_train
        
        dW, db, _ = affine_backward(error, X, W, b)
        
        W -= lr * dW
        b -= lr * db
        
        if verbose and ep % 100 == 0:
            print ('epoch', ep, '/', epochs, ': loss = ', loss)
            loss_history.append(loss)
            
    print('Pelatihan Selesai Bos')
    return W,b, loss_history
```

```python
W, b, loss = train_one_layer(x_train, y_train_ori, epochs = 2000, lr=0.005)
```
